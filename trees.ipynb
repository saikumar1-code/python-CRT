```python
class stack:
    def __init__(self):
        self.l=[]
    def push(self,a):
        self.l.append(a)
    def pop(self):
        self.l.pop()
    def display(self):
        b="->".join(map(str,self.l))
        print(b)
    def peek(self):
        print(self.l[-1])

s=stack()
s.push(100)
s.push(90)
s.push(300)
s.display()
s.pop()
s.display()
s.peek()
```
