```python
def square():
    side = float(input("Enter side: "))
    area = side * side
    print("Area of square =", area)

def rectangle():
    l = float(input("Enter length: "))
    b = float(input("Enter breadth: "))
    area = l * b
    print("Area of rectangle =", area)

def circle():
    r = float(input("Enter radius: "))
    area = 3.14 * r * r
    print("Area of circle =", area)

def triangle():
    b = float(input("Enter base: "))
    h = float(input("Enter height: "))
    area = 0.5 * b * h
    print("Area of triangle =", area)

while True:
    print("\n--- MENU ---")
    print("1. Square")
    print("2. Rectangle")
    print("3. Circle")
    print("4. Triangle")
    print("5. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        square()
    elif choice == 2:
        rectangle()
    elif choice == 3:
        circle()
    elif choice == 4:
        triangle()
    elif choice == 5:
        print("Exiting program...")
        break
    else:
        print("Invalid choice")
```

```python
def prime(n):
    if n<=1:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False
        return True

def sod(n):
    d=[int(i) for i in str(n)]
    return sum(d)

def every(n):
    for i in range(3):
        d=n%10
        n=n//10
        if not prime(d):
            return False
    return True

for i in range(100,1000):
    if prime(i) and prime(sod(i)) and every(i):
        print(i,end=" ")
```
