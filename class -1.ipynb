```python
legs=int(input("enter no of legs:"))
heads=int(input("enter no of heads:"))
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4 + hens*2
  if cal_legs==legs:
    print("Cows",cows)
    print("Hens",hens)
    break
```

```python
legs = int(input("Enter no of legs: "))
heads = int(input("Enter no of heads: "))
flag = False
for cows in range(0, heads + 1):
    hens = heads - cows
    cal_legs = cows * 4 + hens * 2
    if cal_legs == legs:
        flag = True
        break
if flag:
    print("Cows:", cows)
    print("Hens:", hens)
else:
    print("No solution")
```

```python
from datetime import date
fromdate = input("Enter date (dd/mm/yyyy): ")
todate = input("Enter date (dd/mm/yyyy): ")

d1, m1, y1 = map(int, fromdate.split(""))
d2, m2, y2 = map(int, todate.split(" "))

# create date objects
a = date(y1, m1, d1)
b = date(y2, m2, d2)

print("No. of days lived:", (b - a).days)
```
