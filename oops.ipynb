```python
def flower(a):
    if a ==4:
        return
    print(a + 10, end=" ")
    flower(a + 1)
    print(a, end=" ")

flower(1)
```

```python
class abc:
    num=100
    def set_dim(self,a,b):
        self.total=a+b+abc.num

    def display(self):
        print("addition is: ",self.total)
a=abc()
b=abc()

a.set_dim(10,20)
b.set_dim(30,40)

a.display()
b.display()
```
